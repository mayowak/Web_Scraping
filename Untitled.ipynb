```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://ng.indeed.com/jobs?q=data+analyst&l="
```

```python
job = input("Job Location? ")
```

```python
response = requests.get(url + job)
```

```python
soup = BeautifulSoup(response.content)
```

```python
soup
```

```python
results = soup.find(id='resultsCol')
```

```python
results
```

```python
jobs = results.find_all('div', class_='result')
jobs = results.find_all('div', class_='row')
len(jobs)
```

```python
jobs[0]
```

```python
title = jobs[0].find('h2')
title
```

```python
title_link = title.find('a')
title_link
```

```python
link_text = title_link.text
link_text
```

```python
link_text.strip()
```

```python
job_titles = [job.find('h2').find('a').text.strip() for job in jobs]
job_titles
```
